A CONTINUACIÓN, COMENZAMOS A DESARROLLAR LAS DIFERENTES AGRUPACIONES Y TRANSFORMACIONES.  

**Tomamos el 21 de septiembre de 2020 como fecha maxima para comparar con las fecha de compra.**

----------------

```python
def min_time(df):
    times=df.Fecha_compra.apply(lambda x:pd.to_datetime("2020-09-21")-x )
    minimo=times.min()
    return minimo
```

```python
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
df1=df_inicial.groupby("Codigo_Cliente").agg({'ID_Contacto':'count','Valor Neto':'sum' }).reset_index()
df2=df_inicial.groupby("Codigo_Cliente").apply(min_time).reset_index()
df_group_inicial=pd.merge(df1,df2, on="Codigo_Cliente")
# Esta celda dura aprox 20 min
```

```python
df_group_inicial.rename(columns={0:"Dias_desde_ultima_compra", 'Valor Neto':'Total_dinero','ID_Contacto':'Numero_de_compras'}, inplace=True)
df_group_inicial.Dias_desde_ultima_compra=df_group_inicial.Dias_desde_ultima_compra.apply(lambda x:x.days)
df_group_inicial.rename(columns={"Numero_de_compras":"ctd_articulos_comprados"}, inplace=True)
```

```python
df_temp=df_inicial.groupby(["Codigo_Cliente","Fecha_compra"]).size().groupby("Codigo_Cliente").size().reset_index().rename(columns={0:"Dias_diferentes"})
df_group_inicial_=pd.merge(df_group_inicial,df_temp, on="Codigo_Cliente")
df_group_inicial_["prom_dinero/visita"]=round(df_group_inicial_["Total_dinero"]/df_group_inicial_["Dias_diferentes"],0)
df_group_inicial_.drop(columns="Dias_diferentes", inplace=True)
```

**CONSTRUIMOS UNA TABLA CON OTRAS VARIABLES DE INTERÉS POR CLIENTE**

------------

```python
df_1=df_inicial.groupby(["Codigo_Cliente","Factura"]).size().reset_index()
df_1=df_1.groupby("Codigo_Cliente").agg({'Factura':'count'}).rename(columns={"Factura":"ctd_facturas"}).reset_index()
df_1["ctd_articulos_comprados"]=df_group_inicial["ctd_articulos_comprados"]
df_1["prom_articulos/factura"]=round(df_1["ctd_articulos_comprados"]/df_1["ctd_facturas"],2)
df_1.drop(columns=["ctd_facturas","ctd_articulos_comprados"], inplace=True)
```

```python
df_1.head(2)
```

```python
df_2 = df_inicial.groupby(['Codigo_Cliente','Factura','Fecha_compra'])['Tipo de Artículo'].size().reset_index().rename(columns={'Tipo de Artículo':'Ctd_Artículos'})
df_2 = df_2.sort_values(['Codigo_Cliente', 'Fecha_compra'])
df_2['Dias_entre_compras'] = df_2.groupby(['Codigo_Cliente'])['Fecha_compra'].diff().dt.days
df_2_final=df_2.groupby(['Codigo_Cliente'])['Dias_entre_compras'].mean().reset_index().rename(columns={'Dias_entre_compras':'Prom_dias_entre_compras'})
df_2_final.fillna(0, inplace=True)
df_2_final["Prom_dias_entre_compras"]=round(df_2_final["Prom_dias_entre_compras"],0)
```

```python
df_2_final.head(2)
```

```python
def rango_tiempo(df):
    maximo=df.Fecha_compra.max()
    minimo=df.Fecha_compra.min()
    rango=maximo-minimo
    return rango.days
```

```python
df_3=df_inicial.groupby('Codigo_Cliente').apply(rango_tiempo).reset_index()
# Esta celda dura aprox 15 min
```

```python
df_3.rename(columns={0:"Rango_1ra_a_ultima_compra (dias)"}, inplace=True)
df_3.head(2)
```

```python
df_4=df_inicial.groupby(["Codigo_Cliente","Fecha_compra"]).size()
df_4=df_4.groupby("Codigo_Cliente").size().reset_index()
df_4.rename(columns={0:'Visitas'},inplace=True)
df_4["Visitas/año_(2 años)"]=df_4.Visitas/2     # PORQUE SON 2 AÑOS
df_4.drop(columns="Visitas", inplace=True)
```

```python
df_4.head(2)
```

Unimos estas 4 tablas

```python
df_group_interes=pd.merge(df_3,df_2_final, on="Codigo_Cliente")
df_group_interes=pd.merge(df_group_interes,df_4, on="Codigo_Cliente")
df_group_interes=pd.merge(df_group_interes,df_1, on="Codigo_Cliente")
```

**REALIZAMOS UNA TABLA PARA CONOCER EN QUE MESES COMPRA CADA CLIENTE:**

-------------

```python
df_mes=df_inicial.copy()
df_mes["Mes_de_compra"]=df_inicial.Fecha_compra.apply(lambda x: x.month)
df_mes["Mes_de_compra"]=df_mes["Mes_de_compra"].map({1:'Enero',2:'Febrero',3:'Marzo',4:'Abril',
                                                     5:'Mayo',6:'Junio',7:'Julio',8:'Agosto',
                                                     9:'Septiembre',10:'Octubre',11:'Noviembre',12:'Diciembre'})
df_mes_=pd.get_dummies(df_mes.iloc[:,[1,31]])
df_group_mes=df_mes_.groupby("Codigo_Cliente").sum().reset_index()
```
